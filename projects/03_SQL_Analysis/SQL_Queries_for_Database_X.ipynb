# SQL Queries for Database X

This notebook contains SQL queries used for analyzing Database X. The queries are designed to extract insights and perform various operations on the data stored in the database.

```python
## Example SQL Queries

# 1. Select all records from the users table
SELECT * FROM users;

# 2. Count the number of orders placed by each user
SELECT user_id, COUNT(order_id) AS order_count
FROM orders
GROUP BY user_id;

# 3. Find the total sales amount for each product
SELECT product_id, SUM(amount) AS total_sales
FROM sales
GROUP BY product_id;

# 4. Get the average order value
SELECT AVG(total_amount) AS average_order_value
FROM orders;

# 5. List users who have not placed any orders
SELECT u.user_id, u.username
FROM users u
LEFT JOIN orders o ON u.user_id = o.user_id
WHERE o.order_id IS NULL;
```
